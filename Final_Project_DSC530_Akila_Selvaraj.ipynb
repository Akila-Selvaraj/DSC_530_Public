```python
# The following function computes the covariance of two variables
def Cov(xs, ys, meanx=None, meany=None):
    xs = np.asarray(xs)
    ys = np.asarray(ys)

    if meanx is None:
        meanx = np.mean(xs)
    if meany is None:
        meany = np.mean(ys)

    cov = np.dot(xs-meanx, ys-meany) / len(xs)
    return cov
```

```python
# Function to calculate Pearson's correlation

def Corr(xs, ys):
    xs = np.asarray(xs)
    ys = np.asarray(ys)

    meanx, varx = thinkstats2.MeanVar(xs)
    meany, vary = thinkstats2.MeanVar(ys)

    corr = Cov(xs, ys, meanx, meany) / np.sqrt(varx * vary)
    return corr
```

```python
# Function to calculate Spearman's correlation
def SpearmanCorr(xs, ys):
    xranks = pd.Series(xs).rank()
    yranks = pd.Series(ys).rank()
    return Corr(xranks, yranks)
```

```python
# This function computes p-value of an observed difference in means
class DiffMeansPermute(thinkstats2.HypothesisTest):

    def TestStatistic(self, data):
        group1, group2 = data
        test_stat = abs(group1.mean() - group2.mean())
        return test_stat

    def MakeModel(self):
        group1, group2 = self.data
        self.n, self.m = len(group1), len(group2)
        self.pool = np.hstack((group1, group2))

    def RunModel(self):
        np.random.shuffle(self.pool)
        data = self.pool[:self.n], self.pool[self.n:]
        return data
```

```python
# Permutation Test 
glucose_level_data = stroke_data_df.avg_glucose_level.values, other_data_df.avg_glucose_level.values
# glucose_level_data = stroke_data_df.avg_glucose_level.values, stroke_data_df.avg_glucose_level.values
bmi_level_data = stroke_data_df.bmi.values, other_data_df.bmi.values
```

```python
ht = DiffMeansPermute(glucose_level_data)
pvalue = ht.PValue()
print('P value: %.20f' % pvalue)
```

```python
# Here's the distrubution of the test statistic (the difference in means) over many simulated samples:
ht.PlotCdf()
thinkplot.Config(xlabel='test statistic',
                   ylabel='CDF')
```

```python
# Under the null hypothesis, we often see differences bigger than the observed difference.
class DiffMeansOneSided(DiffMeansPermute):

    def TestStatistic(self, data):
        group1, group2 = data
        test_stat = group1.mean() - group2.mean()
        return test_stat
```

```python
# If the hypothesis under test is that increase in glucose level leads to stroke, 
# then the appropriate test statistic is the raw difference between stroke data and non-stroke data.
# In that case, the p-value is smaller, because we are testing a more specific hypothesis.
ht = DiffMeansOneSided(glucose_level_data)
pvalue = ht.PValue()
print('P value: %.10f' % pvalue)
```

```python
# Difference in standard deviation
class DiffStdPermute(DiffMeansPermute):

    def TestStatistic(self, data):
        group1, group2 = data
        test_stat = group1.std() - group2.std()
        return test_stat
```

```python
ht = DiffStdPermute(glucose_level_data)
pvalue = ht.PValue()
print('P value: %.10f' % pvalue)
```

```python
# Testing correlation
# To check whether an observed correlation is statistically significant, 
# we can run a permutation test with a different test statistic.
class CorrelationPermute(thinkstats2.HypothesisTest):

    def TestStatistic(self, data):
        xs, ys = data
        test_stat = abs(thinkstats2.Corr(xs, ys))
        return test_stat

    def RunModel(self):
        xs, ys = self.data
        xs = np.random.permutation(xs)
        return xs, ys
```

```python
# Testing correlation between avg_glucose_level and bmi
cleaned = stroke_data_df.dropna(subset=['avg_glucose_level', 'bmi'])
data = cleaned.avg_glucose_level.values, cleaned.bmi.values
ht = CorrelationPermute(data)
pvalue = ht.PValue()
print('P value: %.10f' % pvalue)
```

```python
# Chi-square test of avg glucose level
class AvgGlucoseTest(thinkstats2.HypothesisTest):

    def MakeModel(self):
        stroke_data_df, other_data_df = self.data
        self.n = len(stroke_data_df)
        self.pool = np.hstack((stroke_data_df, other_data_df))

        pmf = thinkstats2.Pmf(self.pool)
        self.values = range(35, 44)
        self.expected_probs = np.array(pmf.Probs(self.values))

    def RunModel(self):
        np.random.shuffle(self.pool)
        data = self.pool[:self.n], self.pool[self.n:]
        return data
    
    def TestStatistic(self, data):
        stroke_data_df, other_data_df = data
        stat = self.ChiSquared(stroke_data_df) + self.ChiSquared(other_data_df)
        return stat

    def ChiSquared(self, lengths):
        hist = thinkstats2.Hist(lengths)
        observed = np.array(hist.Freqs(self.values))
        expected = self.expected_probs * len(lengths)
        stat = sum((observed - expected)**2 / expected)
        return stat
```

```python
#Predicting using multiple variables

formula = 'stroke_risk ~ avg_glucose_level + bmi'
model = smf.logit(formula, data=health_data_df)
results = model.fit()
results.summary()
```

```python
formula = 'stroke_risk ~ avg_glucose_level + bmi + hypertension + heart_disease'
formula = 'stroke_risk ~ avg_glucose_level + bmi + hypertension + heart_disease + C(smoking_status)'

model = smf.logit(formula, data=health_data_df)
results = model.fit()
results.summary()
```

```python
# To make a prediction, we have to extract the exogenous and endogenous variables.
endog = pd.DataFrame(model.endog, columns=[model.endog_names])
exog = pd.DataFrame(model.exog, columns=model.exog_names)
```

```python
# The baseline prediction strategy is to guess if patient would get stroke.
# In that case, we're right almost 4% of the time.

actual = endog['stroke_risk']
baseline = actual.mean()
baseline
```

```python
# If we use the previous model, we can compute the number of predictions we get right.

predict = (results.predict() >= 0.5)
true_pos = predict * actual
true_neg = (1 - predict) * (1 - actual)
sum(true_pos), sum(true_neg)
```

```python
# And the accuracy, which is slightly higher than the baseline.

acc = (sum(true_pos) + sum(true_neg)) / len(actual)
acc
```
